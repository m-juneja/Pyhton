# opening a file : 
file handler = open("filename" , "open mode", "buffering ")
options of "open mode" : 
- "w" : to write the data to the file . If any data is already ther in file , it will be delted and present data will be stored.
- "r" : just to read the file 
- "a" : to append the data in the file
- "w+":to read and write the data in the file . the previous data is lost if you write with w+ option. 
- "r+" : to read and write the data in the file. previous data in the fie is not deleted. 
- "a+" :to append the data in the file. if no file is created or there in directory , it will create a new file. 

```python
f=open("myfile.txt","a")
str1=input("Enter the text : ")
f.write(str1)
f.close()
```

```python
f=open("myfile1.xlsx",'a+')
print("Enter text (@ at the end )")
while str1!='@':
    str1=input()
    if(str1!='@'):
        f.write(str1+"\n")
f.close()
```
